```python
img1 = cv2.imread("DIP_Lab_task_01 .png")
img2 = cv2.imread("DIP_Lab_task_02.jpg")
img3 = cv2.imread("DIP_Lab_task_03.png")
img4 = cv2.imread("DIP_Lab_task_04.jpg")
img5 = cv2.imread("DIP_Lab_task_05.jpg")
```

# Task 01

### image dimensions

```python
height, width, channels = img1.shape
print(f'Image Dimensions: {width}x{height} pixels')
```

### frequency distribution of intensities

```python

gray_img = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)


hist = cv2.calcHist([gray_img], [0], None, [256], [0, 256])

plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency')
plt.plot(hist)
plt.xlim([0, 256])
plt.show()
```

### Calculate the range of intensities

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(gray_img)
print(f'Intensity Range: {min_val} to {max_val}')
```

# Task 02

### compare a sub-region of the image by intensity values

```python
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
# taking random values
x, y = 40, 50

sub_rgb = img2[y:y+5, x:x+5]
sub_gray = gray_img2[y:y+5, x:x+5]

print("RGB Sub-region Intensity Values:")
print(sub_rgb)

print("\nGrayscale Sub-region Intensity Values:")
print(sub_gray)
```

# Task 3

###  Identify the noise

```python
gray_image = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

plt.figure()
plt.title('Original Grayscale Image')
plt.imshow(gray_image, cmap='gray')
plt.show()

histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Intensity Value')
plt.ylabel('Frequency')
plt.plot(histogram)
plt.xlim([0, 256])
plt.show()
```

###  smoothing with appropriate kernel

```python
smoothed_image = cv2.medianBlur(gray_image, 5)
plt.figure()
plt.title('Smoothed Image')
plt.imshow(smoothed_image, cmap='gray')
plt.show()
```

#  Task 4


###  reduce the brightness

```python
reduced_img = cv2.subtract(img4, np.array([100,100,100]))
fig, axes = plt.subplots(1, 2, figsize=(12, 6)) 

axes[0].imshow(img4) 
axes[0].set_title('Original Image') 
axes[0].axis('off') 

axes[1].imshow(reduced_img) 
axes[1].set_title('Reduced Brightness Image') 
axes[1].axis('off')
```

# Task 5

### Solve the segmentation problem

```python

rgb_img5 = cv2.cvtColor(img5, cv2.COLOR_BGR2RGB)


_, thresholded_image = cv2.threshold(rgb_img5, 150, 255, cv2.THRESH_BINARY)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(rgb_img5, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Thresholded Image')
plt.imshow(thresholded_image, cmap='gray')
plt.show()
```
